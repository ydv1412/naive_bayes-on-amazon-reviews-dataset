```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import model_selection
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.metrics import log_loss
```

```python
import sqlite3
conn = sqlite3.connect("../amazon-fine-food-review/final.sqlite")
```

```python
data_set = pd.read_sql_query("SELECT * FROM Reviews",conn)
```

```python
data_set.head()
```

```python
cleaned_text = data_set["Cleaned_text"]
score = data_set["Score"]
y = score
```

```python
trigram = CountVectorizer(lowercase = True ,ngram_range = (1,3))
trigram_vector = trigram.fit_transform(cleaned_text)
x = trigram_vector
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split( x , y ,test_size = 0.3 , random_state = 1)
```

```python
clf = BernoulliNB()
clf.fit(x_train,y_train)
```

```python
alpha = 0
BernoulliNB(alpha = alpha, binarize = 0.0 ,fit_prior = True , class_prior = None)
predicted_score = clf.predict(x_test)
accuracy = accuracy_score(y_test , predicted_score , normalize = True)*float(100)
print("for alpha = %d accuracy is %d "%(alpha,accuracy))
```

```python
for i in range(len(data_set)):
    BernoulliNB(alpha = i , binarize = 0.0 , fit_prior = False , class_prior = None)
    predicted_score = clf.predict(x_test)
    accuracy = accuracy_score( y_test , predicted_score , normalize = True)*100
    print("accuracy for alpha = %d is %d"%(i,accuracy))
```

```python
BernoulliNB(alpha = 1 , binarize = 0.0 , fit_prior = False , class_prior = None)
predicted_score = clf.predict(x_test)
accuracy = accuracy_score( y_test , predicted_score , normalize = True)*100
print("accuracy for alpha = %d is %d"%(i,accuracy))
```
